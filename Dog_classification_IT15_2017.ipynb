# 1. Binary klasifikacija

Binarna klasifikacija je klasifikacija čiji je zadatak da doda jednu od dve klasne labele. Najbolji primeri za to su:

*  Klasifikacija mejlova (spam ili not spam),
*  Predikcija konverzije (kupiti ili ne).

Binarna klasifikacija se koristi u većini slučajeva tako da postoji klasa koja je glavna - normalna klasa i ona koja je abnormalna. Na primeru spam mejla, mejl koji nije spam bi bio normalna klasa, a spam bi se tretirao kao abnormalna klasa.

Popularni algoritmi koji se koriste za binarnu klasifikaciju su: Logistic regression, k-Nearest Neighbours, Decision Trees, Support Vector Machine, Naive Bayes.

Na slici je prikazan *scater plot* binarne klasifikacije podataka.

image.png

# 2.  Multi-Class klasifikacija

Multi-Class klasifikacija je klasifikacija čiji je zadatak da doda jednu od više mogućih klasnih labela. Primeri za to su:

*  Klasifikacija lica,
*  Klasifikacija vrsta biljaka.

Razlika između Multi-Class klasifikacije i binarne klasifikacije jeste ta što Multi-Class klasifikacija ne podržava postojanje normalne i abnormalne klase već primeri mogu biti klasifikovani jednom od labela klasa iz većeg broja postojećih labela klasa.

Popularni alogritmi koji se koriste za Multi-Class klasifikaciju su: k-Nearest Neighbours, Decision Trees, Naive Bayes, Random Forest, Gradient Boosting.

Na slici je prikazan *scater plot* za Multi-Class klasifikaciju.

# 3. Multi-Label klasifikacija

Multi-Label klasifikacija je klasifikacija gde postoji mogućnost da jedan primer dobije više klasnih labela. Najbolji primer bi bio u korišćenju klasifikacije fotografija - na fotografiji se može naći veći broj objekata te fotografija može dobiti više labela: "jabuka", "sto", "osoba".

Za Multi-Label klasifikaciju koristi se više binarnih klasifikacija kako bi se utvrdila predikcija svake određene klase.

Popularni algoritmi koji se koriste za Multi-Label klasifikaciju su modifikovani algoritmi koji se koriste za binarnu i Multi-Class klasifikaciju kao što su: Multi-label Desicion Trees, Multi-label Random Forests, Multi-label Gradient Boosting.

# 4. Imbalanced klasifikacija

Imbalanced klasifikacija je klasifikacija gde klasifikovani primeri nisu jednako raspoređeni po klasama već jedna klasa ima mnogo veći broj primera od druge. Primeri za Imbalanced klasifikaciju su:

*  Detekcija prevare
*  Detekcija odstupanja
*  Medicinski dijagnostički testovi

Za Imbalanced klasifikaciju se koriste algoritmi binarne klasifikacije gde bi većina primera pripadala normalnoj klasi, ali se mogu koristiti i specifične verzije drugih algoritama kao što su: Random Undersampling, SMOTE Oversampling.

Pre samog kreiranja modela, potrebno je definisati veličinu slika. Takođe, za lakše snalaženje, definisane su putanje do foldera sa slikama. U projektu su dobijena dva foldera gde bi se jedan folder koristio za treniranje, a drugi za testiranje.

Za početak projekta je dobijen i excel fajl labels.csv u kojem su izlistane rase pasa.

```python
labels = pd.read_csv('/content/labels.csv')
labels.head()
sample_df = pd.read_csv('/content/sample_submission.csv')

```

Narednom naredbom je dobijen ukupan broj rasa.

```python
#Kreira se lista svih postojećih rasa sortirana po alfabetu
classes = sorted(list(set(labels['breed'])))
n_classes = len(classes)
#Kreira se dictionary gde se mapira naziv klase sa odgovarajucim brojem
class_to_num = dict(zip(classes, range(n_classes)))
class_to_num
```

Za dalji rad potrebno je kreirati niz od slika. Za to će se koristiti funkcija *images_to_array*.

Niz slika bi se čitao iz foldera. Očitane slike bi se resize-ovale i ubacile u niz. Jedan niz bi se koristio za čuvanje piksela slike dok je drugi korišćen za čuvanje rase psa sa date slike.

Funkcija *to_categorical()* služi za konvertovanje *numpy* niza u binarni niz za dalje korišćenje.

```python
input_shape = (224,224,3)
def images_to_array(data_dir, labels_dataframe, img_size = input_shape):
    #Naziv slika iz csv fajla
    images_names = labels_dataframe['id']
    #Naziv rasa iz csv fajla
    images_labels = labels_dataframe['breed']
    data_size = len(images_names)
    #Kreiaju se dva niza
    X = np.zeros([data_size, img_size[0], img_size[1], img_size[2]], dtype=np.uint8)
    y = np.zeros([data_size,1], dtype=np.uint8)

    #Nizovi se popunjavaju, X je skup piksela date slike dok je y za čuvanje rase psa sa slike
    for i in tqdm(range(data_size)):
        image_name = images_names[i]
        img_dir = os.path.join(data_dir, image_name+'.jpg')
        img_pixels = load_img(img_dir, target_size=img_size)
        X[i] = img_pixels
        
        image_breed = images_labels[i]
        y[i] = class_to_num[image_breed]
    
    #Konvertovanje niza u binarni niz
    y = to_categorical(y)
    return X, y
```

U narednoj liniji se kreiraju nizovi X i y pomoću prethodno definisane funkcije.

```python
X,y = images_to_array(train_path, labels[:])
```

Naredni korak bi bio da se kreira funkcija za ektraktovanje *feature*-a iz dobijenog dataseta pomoću kreiranog modela za treniranje.

Funkcija koja će se koristiti za to je *get_features*. U datoj funkciji se koriste funkcije iz *keras.layers* biblioteke.

*Input* se koristi za dodavanje ulaznih parametara za dalje kreiranje modela.
*Lambda* se koristi za kreiranje *layer*-a od proizvoljnih izraza i predprocesiranje ulaznih parametara.

Za kreiranje *output*-a, koristi se *GlobalAveragePooling2D*. To je operacija koja zamenjuje sve povezane layer-e sa klasičnim konvolucionim neuronskim mrežama. Ideja jeste da se generiše mapa za svaku odgovarajuću kategoriju iz zadatka. Umesto dodavanja potpuno povezane slojeve na mape, uzimamo prosek svake mape, a rezultujući vektor se ubacuje direktno u *softmax* sloj.

```python
img_size = (224,224,3)
def get_features(model_name, data_preprocessor, input_size, data):

    #Input
    input_layer = Input(input_size)
    preprocessor = Lambda(data_preprocessor)(input_layer)
    base_model = model_name(weights='imagenet', include_top=False,
                            input_shape=input_size)(preprocessor)
    #Output
    avg = GlobalAveragePooling2D()(base_model)
    feature_extractor = Model(inputs = input_layer, outputs = avg)
    
    #Ekstraktovanje feature-a sa slika
    feature_maps = feature_extractor.predict(data, batch_size=64, verbose=1)
    print('Feature maps shape: ', feature_maps.shape)
    return feature_maps
```

Za prepoznavanje slika odlučeno je da će se koristiti *Inception v3*. To je model koji se pokazao sa dobrom uspešnošću prepoznavanja slika sa 78.1% uspešnosti sa ImageNet dataset-ovima.

Model se sastoji od simetričnih i asimetričnih blokova i koristiće se za predprocesiranje modela.
image.png

```python
from keras.applications.inception_v3 import InceptionV3, preprocess_input
inception_preprocessor = preprocess_input
inception_features = get_features(InceptionV3,
                                  inception_preprocessor,
                                  img_size, X)
```

```python
final_features = inception_features
print('Final feature maps shape', final_features.shape)
```

Sledeći korak jeste kreiranje duboke neuronske mreže. Duboke neuronske mreže se masovno koriste jer imaju nekoliko slojeva predikcije. Svaki sloj čuva podatke o ulazu i dobijenom izlazu. Nije uvek dobro imati što više slojeva kako bi slojevi obrađivali ulazne podatke na određeni način u zavisnosti od cilja odnosno šta želimo da dobijemo kao rezultat.
Duboke neuronske mreže imaju minimum dva skrivena sloja kao što je prikazano na slici.




image.png

Za kreiranje DNN (*Deep Neural Network*) koristiće se *Sequential* model. On je pogodan za kreiranje mreže kada postoji tačno jedan ulazni parametar i tačno jedan izlaz. U datom slučaju, želimo da, na osnovu jedne unete slike (jednog ulaza), dobijemo rasu psa sa slike (jedan izlaz).

```python
dnn = keras.models.Sequential([
    InputLayer(final_features.shape[1:]),
    Dropout(0.7),
    Dense(n_classes, activation='softmax')
])

#Komplajliranje modela
dnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

#Treniranje modela
h = dnn.fit(final_features, y,
            batch_size=128,
            epochs=60,
            validation_split=0.1)
```

```python
# plot the loss
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(h.history['accuracy'], label='train acc')
plt.plot(h.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```

Kako je treniranje modela završeno sa 91% uspešnosti, potrebno je ponovoviti sve korake i za testiranje modela.

```python
def images_to_array_test(data_dir, labels_dataframe, img_size = input_shape):

    images_names = labels_dataframe['id']
    data_size = len(images_names)
    X = np.zeros([data_size, img_size[0], img_size[1], 3], dtype=np.uint8)
    
    for i in tqdm(range(data_size)):
        image_name = images_names[i]
        img_dir = os.path.join(data_dir, image_name+'.jpg')
        img_pixels = tf.keras.preprocessing.image.load_img(img_dir, target_size=img_size)
        X[i] = img_pixels
        
    print('Ouptut Data Size: ', X.shape)
    return X

test_data = images_to_array_test(test_path, sample_df)
```

```python
#Extract test data features.
inception_features = get_features(InceptionV3, inception_preprocessor, img_size, test_data)

test_features = inception_features
print('Final feature maps shape', test_features.shape)
```
